# 데이터 분석과 시각화

## 1. 시각화와 그래프

### 01. 시각화의 두 가지 목적

시각화의 중요성은 아래와 같다.

1. 분석에 도움을 준다.
2. 보이지 않던 문제, 다시 말해 이상치(Outlier)를 발견하고 제거하는데 도움을 준다.
3. 리포팅에 도움을 준다.

### 02. 선 그래프

어떤 변화를 보여주기에 가장 적합한 그래프다.

이때 선 그래프는 보통 X축에는 시간의 값을 입력하고 Y축에는 해당 시간에 따른 변화를 통해 관찰하고자 하는 값을 입력한다. 따라서 수치화된 자료의 연속된 변화를 파악하는데 유리하기 때문에 문자열 등의 데이터를 이용하는데는 뒤이어 배울 막대 그래프가 훨씬 유용하다.

Pandas의 데이터프레임과 시리즈의 경우 모두 기본적으로 `plot` 메서드를 내장하고 있다. 이때 `kind` 매개변수의 `"line"` 값을 전달해줘야 선 그래프가 그려지는데 이는 매개변수의 기본 값이다. 또한 `y` 매개변수의 값으로 표현하고자 하는 Y축의 값을 넘겨

### 03. 국가별 경제 성장

```Python
%matplotlib inline
import pandas as pd


df = pd.read_csv('data/gdp.csv', index_col=0)

countries: list[str] = [
    "Korea_Rep",
    "United_States",
    "United_Kingdom",
    "Germany",
    "China",
    "Japan"
]
df.plot(y=countries)
```

### 05. 막대 그래프

다양한 카테고리를 비교하기 위해 사용된다.

`kind` 매개변수의 값으로 `"bar"` 값을 전달해야 막대 그래프가 그려지며 이때 `"barh"` 값을 전달할 경우 축이 바뀐다. 또한 `stacked` 매개변수의 값으로 `True` 값을 전달할 경우 비교하고자 하는 카테고리가 하나의 막대 위에 누적되어 보여진다.
